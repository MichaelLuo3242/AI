##  缺失值、异常值

```python
data = pd.read_csv(r'D:\data\hz_weather.csv')
data.head()
```

```python
#查看data的缺失值情况
data.isnull().sum()
```

```python
#建立数据透视表,要求行索引为天气,列索引为风向,对最高气温求均值,并把生成的数据透视表赋值给data2
data2 = pd.pivot_table(data,index="天气",columns="风向",values="最高气温")
data2
```

```python
#查看data2的缺失值情况
data2.isnull().sum()
```

```python
data2.isnull().mean()
```

删除空值
可使用函数df.dropna,默认并不会改变原始数据

```python
#删除缺失值,要求保留完整的行信息
data2.dropna()
```

```python
#删除缺失值,要求保留完整的列信息
data2.dropna(axis=1)
```

缺失值的填充

```python
# 使用特定字符串(missing)填充缺失值
data2.fillna("missing")
```

```python
# 使用前一条记录的数据来代替缺失值
data2.fillna(method='ffill')
```

```python
# 要求使用前一条记录的数据填充缺失值,最大填充次数为1
data2.fillna(method='ffill',limit=1)
```

```python
# 用均值对缺失值进行填充
data2.fillna(data2.mean())
```

查找异常值

```python
data.head()
```

```python
#规定异常值的取值范围,上限h为最高气温的均值+最高气温的1倍标准差,下限为最高气温的均值-最高气温的1倍标准差
h = data.最高气温.mean()+data.最高气温.std()
h
```

```python
l = data.最高气温.mean()-data.最高气温.std()
l
```

```python
#在原数据集中增加一列‘异常值’,如果最高气温低于下限l或者高于上限h则为异常值
data["异常值"] = (data["最高气温"] > h)|(data["最高气温"] < l)
```

```python
#查看"异常值"列中分类情况以及每种分类有多少条记录
data["异常值"].value_counts()
```

```python
#查看异常的数据有哪些
data[data["异常值"] == True]
```
